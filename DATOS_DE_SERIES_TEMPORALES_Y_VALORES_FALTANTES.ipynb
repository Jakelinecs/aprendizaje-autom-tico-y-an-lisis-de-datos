```python
#Creación de datos de series temporales


import numpy as np
import pandas as pd
dates = pd.date_range(start="2021-01-01", end="2021-12-31")
dates
```

```python
df = pd.DataFrame(np.random.randint(1, 100, 365), index=dates, columns=["ランダムな値"])
df
```

```python
#Agrupación de datos


df.groupby(pd.Grouper(freq='M')).mean()
```

```python
df.resample('M').mean()
```

```python
#Imputación de valores faltantes


#Compruebe si faltan valores
df = pd.read_csv("sales_data_contains_nan.csv")
df
```

```python
#Eliminar filas con valores faltantes

dropna_df = df.dropna()
dropna_df
```

```python
#Imputar valores faltantes con un valor especificado


fillna_df = df.fillna(0)
fillna_df
```

```python
ffill_df = df.fillna(method = 'ffill')
ffill_df
```

```python
#Calculado utilizando la media, la mediana y la moda.


fill_mean_df = df.fillna(df.mean(numeric_only=True))
fill_mean_df
```

```python
df.mode()
```

```python
fill_freq_df = df.fillna(df.mode().iloc[0, :])
fill_freq_df
```
