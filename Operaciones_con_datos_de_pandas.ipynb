```python
import numpy as np
import pandas as pd

#Leer los datos del archivo

df = pd.read_csv("sales_data.csv")
df
```

```python
#Extracción de datos mediante la especificación de condiciones
df["売り上げ"] >= 500000
```

```python
matched_df = df[df["売り上げ"] >= 500000]
matched_df
```

```python
#El método de consulta

df.query('売り上げ >= 500000')
```

```python
df.query('客数 >= 150 and 客単価 >= 3500')
```

```python
#Conversión de tipos de datos

df.dtypes
```

```python
df['日付'] = pd.to_datetime(df['日付'])
```

```python
df['客単価'] = df['客単価'].astype(np.float32)
```

```python
#Ordenar datos

df.sort_values(by="売り上げ")
```

```python
df.sort_values(by="売り上げ", ascending=False)
```

```python
#Eliminar una columna


df = df.drop("客単価", axis=1)
df
```

```python
df = df.drop(0, axis=0)
```

```python
df
```

```python
#Cálculos columna a columna


df.loc[:, "客単価"] = df.loc[:, "売り上げ"] / df.loc[:, "客数"]
df
```

```python
df.loc[:, "客単価"] = df.loc[:, "客単価"].round().astype(int)
df
```

```python
#8. Aplicación de funciones mediante el método apply


df["割引後客単価"] = df["客単価"].apply(lambda x: x * 0.9)
df
```

```python
def categorize_unit_price(x):
    if x >= 4000:
        return "高価格帯"
    elif x >= 3000:
        return "中価格帯"
    else:
        return "低価格帯"

df["価格帯"] = df["客単価"].apply(categorize_unit_price)
df
```

```python
df.select_dtypes(include=np.number).apply(np.sum, axis=0)
```
