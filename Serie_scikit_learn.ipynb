```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(data0[:, 0], data0[:, 1], marker="o", label='target 0')
ax.scatter(data1[:, 0], data1[:, 1], marker="x", label='target 1')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```

```python
#########################3   Aprendizaje                   ####


from sklearn.svm import SVC


def plot_boundary_margin_sv(data0, target0, data1, target1, kernel, C, xmin=-1, xmax=1, ymin=-1, ymax=1):
    svc = SVC(kernel=kernel, C=C)  # <-- 1
    svc.fit(np.vstack((data0, data1)), np.hstack((target0, target1)))  # <-- 2
    # 3
    fig, ax = plt.subplots()
    ax.scatter(data0[:, 0], data0[:, 1], marker="o", label='target 0')
    ax.scatter(data1[:, 0], data1[:, 1], marker="x", label='target 1')
    # 4
    xx, yy = np.meshgrid(np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100))
    xy = np.vstack([xx.ravel(), yy.ravel()]).T
    p = svc.decision_function(xy).reshape((100, 100))
    ax.contour(xx, yy, p, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=[':', '-', ':'])
    ax.scatter(svc.support_vectors_[:, 0], svc.support_vectors_[:, 1], s=250, facecolors='none', edgecolors='red') # <-- 5

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc='best')
    plt.show()



```

```python

#Creaci√≥n de una instancia de una m√°quina de vectores de soporte
#Ejecuci√≥n de la formaci√≥n
#C√≥mo trazar un diagrama de dispersi√≥n
#Grafique el l√≠mite de decisi√≥n y los m√°rgenes
#Marcado de vectores de soporte


#Ahora, ejecutemos la funci√≥n definida y grafiquemos los resultados del entrenamiento.

plot_boundary_margin_sv(data0, target0, data1, target1, kernel='linear', C=1e6)
```

```python
#CPuedes ampliar o reducir el margen modificando el valor especificado como argumento al llamar a la funci√≥n .
#CSi reduces el valor de, el margen se ampliar√°.
#Intenta Csustituir 0.1y ejecutar la funci√≥n.

plot_boundary_margin_sv(data0, target0, data1, target1, kernel='linear', C=0.1)
```

```python
######################################################################################33
##                                     ¬øQu√© es un √°rbol de decisiones?

'''

Al utilizar un √°rbol de decisi√≥n para la clasificaci√≥n, es necesario decidir qu√© caracter√≠stica utilizar como punto de partida para la divisi√≥n.
Al decidir las condiciones de la divisi√≥n, es importante considerar la ganancia de informaci√≥n
. Cuanto mayor sea la precisi√≥n de la clasificaci√≥n
, mayor ser√° la ganancia. La precisi√≥n de la clasificaci√≥n se determina por la impureza (la proporci√≥n de elementos innecesarios en una clase), y cuanto menor sea esta impureza, m√°s precisa ser√° la clasificaci√≥n.

La ganancia de informaci√≥n explicada anteriormente se puede calcular utilizando la siguiente f√≥rmula:

Ganancia de informaci√≥n = impureza del nodo padre - suma de impurezas de los nodos secundarios

Cuando la ganancia de informaci√≥n es positiva: la impureza del nodo padre es mayor que la suma de las impurezas de los nodos secundarios, por lo que la divisi√≥n en nodos secundarios reduce la impureza.
Cuando la ganancia de informaci√≥n es negativa: la impureza del nodo padre es menor que la suma de las impurezas de los nodos secundarios, por lo que no dividir en nodos secundarios reduce la impureza.
Existen varios √≠ndices para medir la impureza, como el √≠ndice de impureza de Gini , la entrop√≠a y el error de clasificaci√≥n
. Aqu√≠ explicaremos el √≠ndice de impureza de Gini utilizado por defecto en scikit-learn.
'''




```

```python
###########################################################################################333

'''
√çndice de evaluaci√≥n de la precisi√≥n de la clasificaci√≥n de categor√≠as
Al evaluar la precisi√≥n de un modelo de clasificaci√≥n creado mediante la predicci√≥n de la clasificaci√≥n de datos, las m√©tricas utilizadas difieren entre la clasificaci√≥n binaria y la clasificaci√≥n multiclase.
A continuaci√≥n, explicaremos las m√©tricas b√°sicas de evaluaci√≥n para la clasificaci√≥n binaria. Existen numerosas m√©tricas
para evaluar la precisi√≥n de un modelo, como la precisi√≥n , la recuperaci√≥n , la medida F y el √≠ndice de precisi√≥n . Para calcular las cuatro m√©tricas mencionadas, se suele utilizar una tabla denominada matriz de confusi√≥n
, como se muestra a continuaci√≥n.

 	Resultados positivos	historial negativo
La predicci√≥n es positiva	tp (predicho como un ejemplo positivo y realmente positivo)	fp (predicho como un ejemplo positivo pero en realidad un ejemplo negativo)
La predicci√≥n es negativa	fn (predicho como un ejemplo negativo pero en realidad un ejemplo positivo)	tn (predicho como ejemplo negativo y ejemplo realmente negativo)
Un ejemplo positivo es aquel que representa la verdad. Por
otro lado, un ejemplo negativo es aquel que representa la falsedad.
Por ejemplo, al predecir la disposici√≥n de un cliente a comprar un producto o servicio, se busca predecir qu√© clientes lo comprar√°n; si lo compran, ser√° un ejemplo positivo, y si no lo compran, ser√° un ejemplo negativo.

Utilizando la matriz de confusi√≥n, podemos calcular los valores de cada √≠ndice de la siguiente manera:'''
```

```python
#Tasa de respuestas correctas


from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
data, target = iris.data[:100, :], iris.target[:100]
data_train, data_test, target_train, target_test = train_test_split(data, target, random_state=123, test_size=0.2)

# Aprendizaje

from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(data_train, target_train)
target_pred = tree.predict(data_test)

#Evaluaci√≥n de la precisi√≥n de la clasificaci√≥n

from sklearn.metrics import classification_report
print(classification_report(target_test, target_pred))






```

```python
##Validaci√≥n cruzada


from sklearn.model_selection import cross_val_score

cross_val_score(tree, data, target, cv=10, scoring='accuracy')



```
