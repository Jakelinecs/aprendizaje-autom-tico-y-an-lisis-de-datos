```python
#################################################################
#Qué es una máquina de vectores de soporte?

#Preparación del conjunto de datos



import numpy as np
import pandas as pd
np.random.seed(123)
df1 = pd.DataFrame({"x": np.random.uniform(size=50), "y": np.random.uniform(size=50), "target": np.repeat(0, 50)})
df2 = pd.DataFrame({"x": np.random.uniform(-1, 0, size=50), "y": np.random.uniform(-1, 0, size=50), "target": np.repeat(1, 50)})
df = pd.concat([df1, df2], ignore_index=True)
df

```

```python
data0 = df.query('target == 0').loc[:, ["x", "y"]].values
target0 = df.query('target == 0').loc[:, "target"].values
data1 = df.query('target == 1').loc[:, ["x", "y"]].values
target1 = df.query('target == 1').loc[:, "target"].values
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(data0[:, 0], data0[:, 1], marker="o", label='target 0')
ax.scatter(data1[:, 0], data1[:, 1], marker="x", label='target 1')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```

```python
#########################3   Aprendizaje                   ####


from sklearn.svm import SVC


def plot_boundary_margin_sv(data0, target0, data1, target1, kernel, C, xmin=-1, xmax=1, ymin=-1, ymax=1):
    svc = SVC(kernel=kernel, C=C)  # <-- 1
    svc.fit(np.vstack((data0, data1)), np.hstack((target0, target1)))  # <-- 2
    # 3
    fig, ax = plt.subplots()
    ax.scatter(data0[:, 0], data0[:, 1], marker="o", label='target 0')
    ax.scatter(data1[:, 0], data1[:, 1], marker="x", label='target 1')
    # 4
    xx, yy = np.meshgrid(np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100))
    xy = np.vstack([xx.ravel(), yy.ravel()]).T
    p = svc.decision_function(xy).reshape((100, 100))
    ax.contour(xx, yy, p, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=[':', '-', ':'])
    ax.scatter(svc.support_vectors_[:, 0], svc.support_vectors_[:, 1], s=250, facecolors='none', edgecolors='red') # <-- 5

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc='best')
    plt.show()



```

```python

#Creación de una instancia de una máquina de vectores de soporte
#Ejecución de la formación
#Cómo trazar un diagrama de dispersión
#Grafique el límite de decisión y los márgenes
#Marcado de vectores de soporte


#Ahora, ejecutemos la función definida y grafiquemos los resultados del entrenamiento.

plot_boundary_margin_sv(data0, target0, data1, target1, kernel='linear', C=1e6)
```

```python
#CPuedes ampliar o reducir el margen modificando el valor especificado como argumento al llamar a la función .
#CSi reduces el valor de, el margen se ampliará.
#Intenta Csustituir 0.1y ejecutar la función.

plot_boundary_margin_sv(data0, target0, data1, target1, kernel='linear', C=0.1)
```

```python
######################################################################################33
##                                     ¿Qué es un árbol de decisiones?

'''

Al utilizar un árbol de decisión para la clasificación, es necesario decidir qué característica utilizar como punto de partida para la división.
Al decidir las condiciones de la división, es importante considerar la ganancia de información
. Cuanto mayor sea la precisión de la clasificación
, mayor será la ganancia. La precisión de la clasificación se determina por la impureza (la proporción de elementos innecesarios en una clase), y cuanto menor sea esta impureza, más precisa será la clasificación.

La ganancia de información explicada anteriormente se puede calcular utilizando la siguiente fórmula:

Ganancia de información = impureza del nodo padre - suma de impurezas de los nodos secundarios

Cuando la ganancia de información es positiva: la impureza del nodo padre es mayor que la suma de las impurezas de los nodos secundarios, por lo que la división en nodos secundarios reduce la impureza.
Cuando la ganancia de información es negativa: la impureza del nodo padre es menor que la suma de las impurezas de los nodos secundarios, por lo que no dividir en nodos secundarios reduce la impureza.
Existen varios índices para medir la impureza, como el índice de impureza de Gini , la entropía y el error de clasificación
. Aquí explicaremos el índice de impureza de Gini utilizado por defecto en scikit-learn.
'''




```
