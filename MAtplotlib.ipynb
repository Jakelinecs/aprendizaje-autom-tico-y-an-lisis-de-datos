```python
import numpy as np
np.random.seed(123)
x1 = np.random.normal(0, 10, 1000)
x2 = np.random.normal(0, 15, 1000)
x3 = np.random.normal(0, 20, 1000)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
labels = ['x1', 'x2', 'x3']
ax.boxplot((x1, x2, x3), labels=labels)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot((x1, x2, x3), labels=labels, vert=False)
plt.show()
```

```python

#dibujar un gráfico circular
import pandas as pd
df = pd.DataFrame([['テレビ、ラジオや新聞等の広告', 5],
                   ['インターネット検索結果', 15],
                   ['インターネット上の広告', 25],
                   ['SNS広告', 50],
                   ['知人からの紹介', 1],
                   ['その他', 4]], columns = ["媒体", "集計結果（%）"])
df
```

```python
import matplotlib.pyplot as plt

labels = df.loc[:, "媒体"]
rate = df.loc[:, "集計結果（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels=labels)
ax.axis('equal')
plt.show()
```

```python
#cambia posición inicial

import matplotlib.pyplot as plt

labels = df.loc[:, "媒体"]
rate = df.loc[:, "集計結果（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels=labels,startangle=90)
ax.axis('equal')
plt.show()
```

```python
#cambiar orden de elementos

import matplotlib.pyplot as plt

labels = df.loc[:, "媒体"]
rate = df.loc[:, "集計結果（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels=labels, startangle=90, counterclock=False)

ax.axis('equal')
plt.show()
```

```python
# con porcentaje en cada elemento

import matplotlib.pyplot as plt
labels = df.loc[:, "媒体"]
rate = df.loc[:, "集計結果（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels=labels, startangle=90, counterclock=False, autopct='%1.0f%%')

ax.axis('equal')
plt.show()




# 省略
```

```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4]
y1 = [5, 4, 7, 8]
x2 = [1, 2, 3, 4]
y2 = [4, 3, 6, 7]

fig, ax = plt.subplots()
ax.plot(x1, y1, label='data1')
ax.bar(x2, y2, label='data2')
ax.legend()
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({'data1': [3, 2, 1], 'data2': [1, 4, 2]})

import matplotlib.pyplot as plt
df.plot()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3])
y1 = np.array([4, 12, 5])
y2 = np.array([8, 2, 14])
y3 = np.array([10, 6, 8])
fig, ax = plt.subplots()
ax.bar(x, y1+y2+y3, label='y3')
ax.bar(x, y1+y2, label='y2')
ax.bar(x, y1, label='y1')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x1, y1)
ax[1].plot(x2, y2)
plt.show()
```
